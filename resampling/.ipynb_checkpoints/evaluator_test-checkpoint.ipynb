```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from plt_utils import PlotScaledData, PlotRelevance, PlotBins, PlotResampledResults

from scipy.interpolate import PchipInterpolator

def PCHIP_interpolator(y_train):
    q75, q25 = np.percentile(y_train, [75 ,25])
    iqr = q75 - q25
    median = np.median(y_train)
    value = [median - 1.5* iqr,median, median + 1.5 *iqr]
    #value = [median, median + 1.5 *iqr]
    relevances = [1,0,1]
    #relevances = [0,1]
    return PchipInterpolator(value, relevances, extrapolate=False)

df1=pd.read_csv("./data/london_merged.csv")
df1=df1.reset_index()['cnt']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
df1=scaler.fit_transform(np.array(df1).reshape(-1,1)) 

def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return np.array(dataX), np.array(dataY)

time_step  = 4
SPLIT_RATIO = 0.7
def split_data(df):
    ##splitting dataset into train and test split
    training_size=int(len(df1)*SPLIT_RATIO)
    test_size=len(df1)-training_size
    train_data,test_data=df1[0:training_size,:],df1[training_size:len(df1),:]

    # reshape into X_train = [x_t, x_t+1, x_t+2, x_t+3] and y_train = [x_t+4]
    time_step = 4
    print(train_data)
    X_train, y_train = create_dataset(train_data, time_step)
    X_test, y_test = create_dataset(test_data, time_step)
    # reshape into k_x = [x_t, x_t+1, x_t+2, x_t+3, x_t+4]
    k_x = np.c_[X_train,y_train]
    return k_x, X_train, y_train, X_test, y_test

k_x, X_train, y_train, X_test, y_test = split_data(df1)

from scipy.special import factorial
import math
import numpy
def calculate_relevance(y_train, relevance_threshold):
    #Relevance function
    RELEVANCE_THRESH = relevance_threshold

    relevance_function = PCHIP_interpolator(y_train)
    # uses PCHIP instead
    r_y = [relevance_function.__call__(val) for val in y_train]
    print(f"number of nans: {np.count_nonzero(np.isnan(r_y))}")
    
    rel_nan = pd.DataFrame(data={'y': y_train,'rel':r_y,'isnan':np.isnan(r_y)})
    #r_y is nan only for samples at which y is outlier (y >= q75 + 1.5*IQR)
    r_y = numpy.nan_to_num(r_y, copy=False, nan=1)
    print(rel_nan[rel_nan['y']>=0.3572996183206107])
    lst=[]
    rels = 0 # relevant
    coms = 0 # common
    for j in r_y:
        if j < RELEVANCE_THRESH:
            lst.append(0) #common
            coms+=1
        else:
            lst.append(1) #relevant
            rels+=1

    k_y=np.array(lst)

    print(f"{coms} common values, {rels} relevant values")
    
    return RELEVANCE_THRESH, r_y, k_y, relevance_function

relevance_threshold = 0.31
RELEVANCE_THRESH, r_y, k_y, relevance_function = calculate_relevance(y_train, relevance_threshold)

PlotRelevance(y_train, r_y, RELEVANCE_THRESH)
```

```python
eval_params = {
                   "RMSE": {},
                   "RMSERare": {},
                   "Utility": {'B_max':10, 'C_max':11, 'm':12},
                   "CaseWeight": {'m':12},
                   "PrecisionRecall": {}
              }

class Evaluator():
    eval_metrics = []
    def __init__(self, num_exps, eval_params, rel_func, rel_thresh):
        for m in EvaluationMetric.__subclasses__():
            self.num_exps = num_exps
            m_name = m.__name__
            if m_name not in eval_params.keys(): continue
            self.eval_metrics.append(m(eval_params[m_name], rel_func, rel_thresh))
        self.results_df = self.initResultsDf()
    
    def initResultsDf(self):
        cols = []
        for m in self.eval_metrics:
            cols = cols + m.m_names
        cols = [item + suffix for suffix in ['_Train','_Test'] for item in cols]
        cols = ["Exp"]+cols
        return pd.DataFrame(columns=cols) 
            
    def evaluateMetrics(self, exp, train_pred, train_actual, test_pred, test_actual):
        exp_row = {"Exp": exp}
        for e in self.eval_metrics:
            tr = e.evaluate(train_pred, train_actual)
            tr = {key + "_Train": value for key, value in tr.items()}
            ts = e.evaluate(test_pred, test_actual)
            ts = {key + "_Test": value for key, value in ts.items()}
            exp_row = {**exp_row, **tr, **ts}
        self.results_df = pd.concat([self.results_df, pd.DataFrame([exp_row])], ignore_index=True)

def rmse(pred, actual):
    error = np.subtract(pred, actual)
    sqerror= np.sum(np.square(error))/actual.shape[0]
    return np.sqrt(sqerror)
        
class EvaluationMetric:
    def __init__(self):
        pass
    
    def evaluate(self):
        pass

class RMSE(EvaluationMetric):
    m_names = ["RMSE"]
    def __init__(self,params,rel_func,rel_thresh):
        pass

    def evaluate(self, pred, actual):
        return {"RMSE": rmse(pred, actual)}
        
    
class RMSERare(EvaluationMetric):
    m_names = ["RMSERare"]
    def __init__(self, params,rel_func,rel_thresh):
        self.rel_func = rel_func
        self.rel_thresh = rel_thresh
        pass

    def evaluate(self, pred, actual):
        actual_rel = self.rel_func.__call__(actual)
        numpy.nan_to_num(actual_rel, copy=False, nan=1)
        rare_ind = np.argwhere(actual_rel >= rthresh)    
        rmse_rare = rmse(pred[rare_ind].flatten(), actual[rare_ind].flatten())
        return {"RMSERare": rmse_rare}

class Utility(EvaluationMetric):
    m_names = ["Utility","Benefit","Cost"]
    def __init__(self,params,rel_func,rel_thresh):
        self.rel_func = rel_func
        self.rel_thresh = rel_thresh
        self.bmax = params['B_max']
        self.cmax = params['C_max']
        self.mbirel = params['m']
    
    def evaluate(self, pred, actual):
        return {"Utility": 3, "Benefit": 4, "Cost": 5}

class CaseWeight(EvaluationMetric):
    m_names = ["CaseWeight"]
    def __init__(self,params,rel_func,rel_thresh):
        self.rel_func = rel_func
        self.rel_thresh = rel_thresh
        self.mbirel = params['m']
    
    def evaluate(self, pred, actual):
        return {"CaseWeight": 6}

class PrecisionRecall(EvaluationMetric):
    m_names = ["Precision", "Recall"]
    def __init__(self,params,rel_func,rel_thresh):
        pass
        
    def evaluate(self, pred, actual):
        return {"Precision": 7, "Recall": 8}

ev = Evaluator(2,eval_params, relevance_function, RELEVANCE_THRESH)
for ex in range(2):
    ev.evaluateMetrics(ex,1,2,3,4)
print(ev.results_df)
```
